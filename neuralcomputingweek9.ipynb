```python
data = pd.read_csv('XAUUSD_5 Mins_Ask_2024.01.01_2024.06.30.csv').merge(pd.read_csv('XAUUSD_5 Mins_Bid_2024.01.01_2024.06.30.csv'), left_on='Time (UTC)', right_on='Time (UTC)', how='outer')

# Rename columns to meaningful names
data.columns = ['Local time', 'Open_Ask', 'High_Ask', 'Low_Ask', 'Close_Ask', 'Volume_Ask',
                   'Open_Bid', 'High_Bid', 'Low_Bid', 'Close_Bid', 'Volume_Bid']

# Create Volume_Delta and Volume_Delta_abs
data["Volume_Delta"] = data["Volume_Ask"] - data["Volume_Bid"]
data["Volume_Delta_abs"] = (data["Volume_Ask"] - data["Volume_Bid"]).abs()

# Create Open_Delta, High_Delta, Low_Delta, Close_Delta
data["Open_Delta"] = data["Open_Ask"] - data["Open_Bid"]
data["High_Delta"] = data["High_Ask"] - data["High_Bid"]
data["Low_Delta"] = data["Low_Ask"] - data["Low_Bid"]
data["Close_Delta"] = data["Close_Ask"] - data["Close_Bid"]

# Create Local_time_T related columns
data['Local_time_T'] = pd.to_datetime(data['Local time'], utc=True)
data['Local_time_T_shipt_1_Down'] = data['Local_time_T'].shift(1)
data['Local_time_T_Delta'] = data['Local_time_T'] - data['Local_time_T_shipt_1_Down']

# Add 'New_day' and 'New_week' columns to 'data' DataFrame
data['New_day'] = 0
data.loc[data['Local_time_T_Delta'] > '0 days 00:05:00', 'New_day'] = 1
# Special handling for the first row of 'New_day' after re-creating it from scratch
data['Local_time_T_Delta'].iloc[0] = "2 days 01:01:00" # This line sets the first element, which is then used by the next line. This order is critical.
data['New_day'].iloc[0] = 1 # Set 'New_day' to 1 for the very first entry as it marks a new day (and a new week)

data['New_week'] = 0
data.loc[data['Local_time_T_Delta'] > '1 days 00:00:00', 'New_week'] = 1

# Create Y_High_Bid and Y_Low_Ask in data as well
data['Y_High_Bid'] = data['High_Bid']
data['Y_Low_Ask'] = data['Low_Ask']

# Now drop the time-related columns that are not needed in data2, including 'Local time'
data2 = data.drop(['Local time', 'Local_time_T', 'Local_time_T_shipt_1_Down', 'Local_time_T_Delta'], axis=1)
```

```python
data2['Volume_Delta_abs']  = data2['Volume_Delta_abs'] / max_volume_Delta_abs
```

```python
# For the price deltas, the minimum difference is more interesting, therefore we apply inverse normalisation
# data= 1 - (data - data.min()) / (data.max() - data.min()) = (data.max() - data) / (data.max() - data.min())
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Input
from tensorflow.keras.optimizers import Adam
from keras_tuner import RandomSearch

# Define your input shape
timesteps = 50   # Replace with your actual sequence length
features = 10    # Replace with your actual number of features

def build_model(hp):
    model = Sequential()
    # Use Input() instead of passing input_shape to LSTM
    model.add(Input(shape=(timesteps, features)))
    model.add(LSTM(
        units=hp.Int('units_1', min_value=20, max_value=100, step=20),
        return_sequences=True
    ))
    model.add(Dropout(
        hp.Float('dropout', min_value=0.1, max_value=0.5, step=0.1)
    ))
    model.add(LSTM(
        units=hp.Int('units_2', min_value=10, max_value=50, step=10)
    ))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(
        optimizer=Adam(
            learning_rate=hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
        ),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )
    return model

# Create the tuner
tuner = RandomSearch(
    build_model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory='my_dir',
    project_name='lstm_tuning'
)
```
